```python
e = 1.0
mantissa = 1
while 1.0 + e / 2.0 > 1.0:
    e /= 2.0
    mantissa += 1
print(f'episolon: {e}')
print(f'mantissa: {mantissa}')
```

```python
import numpy as np

print(f'official epsilon: {np.finfo(float).eps}\n')

e = np.float16(1.0)
mantissa = 1
while np.float16(1.0) + e / np.float16(2.0) > np.float16(1.0):
    e /= np.float16(2.0)
    mantissa += 1
print(f'episolon: {e}')
print(f'official epsilon: {np.finfo(np.float16).eps}')
print(f'mantissa: {mantissa}')

e = np.float32(1.0)
mantissa = 1
while np.float32(1.0) + e / np.float32(2.0) > np.float32(1.0):
    e /= np.float32(2.0)
    mantissa += 1
print(f'\nepisolon: {e}')
print(f'official epsilon: {np.finfo(np.float32).eps}')
print(f'mantissa: {mantissa}')
```

```python
import matplotlib.pyplot as plt

x = np.linspace(0,10)
plt.subplots(constrained_layout = True)[1].secondary_xaxis(0.5)
plt.plot(x, np.sin(x), color='red')
plt.plot(x, np.cos(x), color='orange')
plt.legend(['sin', 'cos'])
plt.title('Sine and cosine from 0 to 10')
plt.show()
```

```python
def iter_fibonacci(n):
    a, b = 0, 1

    for _ in range(n):
        a, b = b, a + b
    return a

def rk(k):
  return iter_fibonacci(k+1)/iter_fibonacci(k)

from matplotlib.ticker import PercentFormatter

phi = (1+5**0.5)/2

arange = np.arange(12)
fig, ax = plt.subplots()
ax.plot(arange, [abs(rk(i+1)-phi)/phi for i in arange])

ax.yaxis.set_major_formatter(PercentFormatter(1))
ax.legend(['error'])
plt.show()
```
